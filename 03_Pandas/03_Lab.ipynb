```python
# (likes / (likes + dislikes)) formülü üzerinden hesaplanarak like_avg sütununa yazalım

def calculate_like_avg(likes: int, dislike: int) -> float:
    if (likes + dislike) == 0:
        return 0.0
    return likes / (likes + dislike)

df['like_avg'] = df.apply(
    lambda row: calculate_like_avg(row['likes'], row['dislikes']),
    axis=1
)

df.sort_values(by='like_avg', ascending=False).head(10)[['title', 'likes', 'dislikes', 'like_avg']]
```

```python
# (likes / (likes + dislikes)) formülü üzerinden hesaplanarak like_avg sütununa yazalım

def calculate_like_avg(likes: pd.Series, dislike: pd.Series) -> list:
    like_list = list(likes)
    dislike_list = list(dislike)

    combined_list = list(zip(like_list, dislike_list))

    avg_list = []

    for like, dislike in combined_list:
        if (like + dislike) == 0:
            avg_list.append(0.0)
        else:
            avg = like / (like + dislike)
            avg_list.append(round(avg, 2))
    
    return avg_list

df['like_avg'] = calculate_like_avg(likes=df['likes'], dislike=df['dislikes'])

df[['title', 'likes', 'dislikes', 'like_avg']].sort_values(by=['like_avg', 'likes'], ascending=[False, False])
```
