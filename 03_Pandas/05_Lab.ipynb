# Auto Dataset — Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    filepath_or_buffer='data/auto.csv'
)

df.head()
```

```python
df.columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style",
              "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight",
              "engine-type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio",
              "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]

df.head()
```

```python
df.replace(
    to_replace='?',
    value=np.nan,
    inplace=True
) 

df.head()
```

```python
df.isna().sum() # Check for missing values in each column
```

```python
missing_values_df = df.isnull() # Create a DataFrame indicating missing values

missing_values_df.head()
```

```python
for column_name in missing_values_df.columns.tolist():
    print(
        '============================\n'
        f'Column Name: {column_name}\n'
        f'Missing Values Count: {missing_values_df[column_name].value_counts()}'
    )
```

```python
df['num-of-doors'].value_counts() # Check the value counts for 'num-of-doors' column
```

```python
df['num-of-doors'].value_counts().idxmax() # Get the most frequent value for 'num-of-doors' column
```

```python
df['num-of-doors'].value_counts().argmax() # Deprecated method to get the most frequent value
```

```python
df['num-of-doors'] = df['num-of-doors'].replace(
    to_replace=np.nan,
    value=df['num-of-doors'].value_counts().idxmax()
) # Fill missing values in 'num-of-doors' with the most frequent value 

df.head()
```

```python
df['city_l/100km'] = 235 / df['city-mpg'] # Create a new column 'city_l/100km' by converting 'city-mpg' to liters per 100 kilometers
df['highway_l/100km'] = 235 / df['highway-mpg'] # Create a new column 'highway_l/100km' by converting 'highway-mpg' to liters per 100 kilometers

df[['make', 'city-mpg', 'city_l/100km']] # Display the 'make', 'city-mpg', and 'city_l/100km' columns to verify the new columns have been added correctly
```

```python
df[['length', 'width', 'height']].describe() # Get summary statistics for 'length', 'width', and 'height' columns
```

```python
df['length'] = df['length'] / df['length'].max() # Normalize 'length' column by dividing each value by the maximum value in the 'length' column
df['width']  = df['width']  / df['width'].max() # Normalize 'width' column by dividing each value by the maximum value in the 'width' column
df['height'] = df['height'] / df['height'].max() # Normalize 'height' column by dividing each value by the maximum value in the 'height' column

df[['length', 'width', 'height']]
```

```python
dummy_variable_df = pd.get_dummies(df['fuel-type'], dtype='float') # Create dummy variables for 'fuel-type' column and convert them to float type

dummy_variable_df
```
