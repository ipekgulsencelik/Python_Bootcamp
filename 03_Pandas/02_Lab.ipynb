```python
df = pd.read_csv(filepath_or_buffer="data/nba.csv")

df.head()
```

```python
# En değerli takımı bulunuz.
(
    df.groupby(by='Team')
      .agg({
          'Salary': 'sum'
      })
      .sort_values(
          by='Salary',
          ascending=False
      )
      .head(1)
)
```

### Yapılan İşlemler

1. **groupby('Team')**
   - Veriyi takımlara göre gruplar

2. **agg({'Salary': 'sum'})**
   - Her takımın toplam maaşını hesaplar

3. **sort_values(ascending=False)**
   - Toplam maaşa göre büyükten küçüğe sıralar

4. **head(1)**
   - En yüksek maaşa sahip takımı getirir

```python
# Takımlara Göre Ortalama Yaş
df_groupby_team = (
    df.groupby(by='Team')
      .agg({
          'Age': 'mean'
      })
      .sort_values(
          by='Age',
          ascending=False
      )
)

df_groupby_team
```

### Bu Hücre Ne Yapıyor?

- **groupby('Team')**
  - Veriyi takımlara göre gruplar

- **agg({'Age': 'mean'})**
  - Her takımın yaş ortalamasını hesaplar

- **sort_values(ascending=False)**
  - En yaşlı takımı en üste alır

Sonuç:
> Ortalama yaşı en yüksek takım en üstte görünür

```python
df_groupby_team['Age'] = df_groupby_team['Age'].apply(lambda x: round(x, 2))

df_groupby_team
```
