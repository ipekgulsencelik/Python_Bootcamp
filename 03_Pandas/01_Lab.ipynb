## `.loc` vs `.iloc`

> `.loc` **isimle (label),** `.iloc` **sıra numarasıyla (index) seçer.**


🔑 **Temel fark**

| Özellik       | `.loc`             | `.iloc`              |
| ------------- | ------------------ | -------------------- |
| Seçim türü    | **Label (etiket)** | **Index (pozisyon)** |
| Satır         | index adı          | 0-based sıra         |
| Sütun         | sütun adı          | sütun sırası         |
| Okunabilirlik | ⭐⭐⭐⭐⭐       | ⭐⭐⭐             |
| Hata riski    | Düşük              | Daha yüksek          |

---

### `.loc` → İsme göre seçim (Label-based)

````python
df.loc[1, "Movie_Title"]
````

**Ne olur?**
- `1` → **index label**
- `"Movie_Title"` → sütun adı

📌 Yani:
> **Index’i 1 olan satırın Movie_Title değeri**

---

#### Sayısal sütun

````python
df.loc[1, "Num_Reviews"]
````

📌 Bu yüzden `.loc`:
- Okunur
- Güvenlidir
- Gerçek projelerde **tercih edilir**

---

### `.iloc` → Sıraya göre seçim (Position-based)
Aynı veriyi `iloc` ile alalım

````python
df.iloc[1, 0]
````

**Ne olur?**
- `1` → 2. satır (0-based)
- `0` → 1. sütun

---


#### Num_Reviews

### `sort_values()` Nedir?

`sort_values()` bir **DataFrame veya Series** içindeki verileri
belirtilen değerlere göre sıralamak için kullanılır.

- Varsayılan olarak **küçükten büyüğe** sıralar.

````python
(
    df.query('YR_Released > 2000')
      .loc[:, ['Movie_Title', 'YR_Released', 'Rating']]
      .sort_values(by='Rating', ascending=False)
)
````

````python
# 2000'den sonra yayınlanmış rating en çok olan top 10 movie
(
    df.query('YR_Released > 2000')
        .loc[:, ['Movie_Title', 'YR_Released', 'Rating']]
        .sort_values(by='Rating', ascending=False)
        .head(10)
)
````

````python
(
    df[df['YR_Released'] > 2000]
      .loc[:, ['Movie_Title', 'YR_Released', 'Rating']]
      .sort_values(by='Rating', ascending=False)
      .head(10)
)
````

````python
(
    df[df['YR_Released'] > 2000]
      [['Movie_Title', 'YR_Released', 'Rating']]
      .sort_values(by='Rating', ascending=False)
      .head(10)
)
````

````python
(
    df[
        (df['Rating'] > 7.0) & (df['Rating'] < 9.0) 
    ][['Movie_Title', 'Rating', 'Num_Reviews']]
    .sort_values(by='Rating', ascending=False)
)
````

````python
(
    df.loc[
        (df['Rating'] > 7.0) & (df['Rating'] < 9.0),
        ['Movie_Title', 'Rating', 'Num_Reviews']
    ]
    .sort_values(by='Rating', ascending=False)
)
````

````python
(
    df.query('YR_Released > 2000 & Rating < 9.0')
        [['Movie_Title', 'YR_Released', 'Rating']]
        .sort_values(by='Rating', ascending=False)
)
````

**2000 ile 3000 arasında Num Reviews** değerine sahip filmleri filtreler, istenen kolonları seçer ve **Num Reviews** değerine göre azalan şekilde sıralar.

````python
(
    df.query('2000 < Num_Reviews < 3000')
      .loc[:, ['Movie_Title', 'Rating', 'Num_Reviews']]
      .sort_values(by='Num_Reviews', ascending=False)
)
````

#### Açıklama
- `query`: Sayısal aralık filtresi
- `loc`: İstenen kolonları seçmek için
- `sort_values`: Yorum sayısına göre sıralama

`Movie_Title` içinde **King** geçen ve `Rating >= 7.0` olan filmleri filtreleyerek sonuçtan `Movie_Title`, `Rating`, `Num Reviews` kolonlarını seçip `Rating` değerine göre **azalan** (çoktan aza) sıralamak

````python
(
    df.query('Movie_Title.str.contains("King", case=False, na=False) & Rating >= 7.0')
    .loc[:, ['Movie_Title', 'Rating', 'Num_Reviews']]
    .sort_values(by='Rating', ascending=False)
)
````

Parametreler:
- `case=False`: büyük/küçük harf duyarsız arama
- `na=False`: `Movie_Title` içinde `NaN/None` varsa hata vermesin

````python
df[
    df["Movie_Title"].str.contains("King", case=False, na=False) & (df["Rating"] > 7.0)
].loc[:, ["Movie_Title", "Rating", "Num_Reviews"]].sort_values(by="Rating", ascending=False)
````

## GroupBy()

Veri seti içerisinde bulunan spesifik bilgilere göre veri seti grouplamaya yarar.

````python
df.groupby('YR_Released')['Num_Reviews'].sum().sort_values(ascending=False)
````

- Tip: **Series**
- Index: `YR_Released`
- Değer: `Num_Reviews` toplamı

Bunu DataFrame'e çevirmek istersen:
```python
df.groupby('YR_Released')['Num_Reviews'].sum().reset_index()
```

````python
(
    df.groupby('YR_Released')
      .agg({'Num_Reviews': 'sum'})
)
````

- Tip: **DataFrame**
- Kolon adı korunur: `Num_Reviews`

İstersen index'i kolona çevirebilirsin:
```python
df.groupby('YR_Released', as_index=False).agg({'Num_Reviews': 'sum'})
```

## Ne zaman hangisi?

| İhtiyaç | Önerilen |
|---|---|
| Tek kolon, hızlı analiz | Path I |
| Birden fazla metrik/kolon | Path II (`agg`) |
| DataFrame çıktısı lazım | Path II |
| Çizim/grafik için hızlı seri | Path I |

### Bonus — Temiz isimlendirme + sıralama

En çok kullanılan modern stil:
- `as_index=False` (çıktı DataFrame ve `YR_Released` kolonda)
- Named aggregation ile yeni kolon adı
- Çoktan aza sıralama


````python
(
    df.groupby('YR_Released', as_index=False)
      .agg(Num_Reviews_Total=('Num_Reviews', 'sum'))
      .sort_values('Num_Reviews_Total', ascending=False)
)
````

````python
(
    df.groupby('YR_Released', as_index=False)
      .agg(Num_Reviews=('Num_Reviews', 'sum'))
      .query('700000 < Num_Reviews < 1000000')
      .sort_values(by='Num_Reviews', ascending=False)
)
````

Filmleri `YR_Released` sütununa göre gruplayarak, her yıl için ortalama `Rating` değerini hesaplayıp ortalama Rating değeri 5.0 ile 10.0 arasında olan yılları filtreleyiniz. Sonucu ortalama Rating değerine göre çoktan aza sıralayınız.

````python
(
    df.groupby('YR_Released', as_index=False)
      .agg(Avg_Rating=('Rating', 'mean'))
      .query('5.0 < Avg_Rating < 10.0')
      .sort_values('Avg_Rating', ascending=False)
)
````
