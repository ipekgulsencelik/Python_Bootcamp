## `.loc` vs `.iloc`

> `.loc` **isimle (label),** `.iloc` **sƒ±ra numarasƒ±yla (index) se√ßer.**


üîë **Temel fark**

| √ñzellik       | `.loc`             | `.iloc`              |
| ------------- | ------------------ | -------------------- |
| Se√ßim t√ºr√º    | **Label (etiket)** | **Index (pozisyon)** |
| Satƒ±r         | index adƒ±          | 0-based sƒ±ra         |
| S√ºtun         | s√ºtun adƒ±          | s√ºtun sƒ±rasƒ±         |
| Okunabilirlik | ‚≠ê‚≠ê‚≠ê‚≠ê‚≠ê       | ‚≠ê‚≠ê‚≠ê             |
| Hata riski    | D√º≈ü√ºk              | Daha y√ºksek          |

---

### `.loc` ‚Üí ƒ∞sme g√∂re se√ßim (Label-based)

```python
df.loc[1, "Movie_Title"]
```

**Ne olur?**
- `1` ‚Üí **index label**
- `"Movie_Title"` ‚Üí s√ºtun adƒ±

üìå Yani:
> **Index‚Äôi 1 olan satƒ±rƒ±n Movie_Title deƒüeri**

---

#### Sayƒ±sal s√ºtun

```python
df.loc[1, "Num_Reviews"]
```

üìå Bu y√ºzden `.loc`:
- Okunur
- G√ºvenlidir
- Ger√ßek projelerde **tercih edilir**

---

### `.iloc` ‚Üí Sƒ±raya g√∂re se√ßim (Position-based)
Aynƒ± veriyi `iloc` ile alalƒ±m

```python
df.iloc[1, 0]
```

**Ne olur?**
- `1` ‚Üí 2. satƒ±r (0-based)
- `0` ‚Üí 1. s√ºtun

---


#### Num_Reviews

```python
df.iloc[1, 3]
```

‚ö†Ô∏è Problem:
- S√ºtun sƒ±rasƒ± deƒüi≈üirse kod **bozulur**
- Okunmasƒ± zor

#### Kar≈üƒ±la≈ütƒ±rma

| Ama√ß      | `.loc`                     | `.iloc`         |
| --------- | -------------------------- | --------------- |
| Film adƒ±  | `df.loc[1, "Movie_Title"]` | `df.iloc[1, 0]` |
| Oy sayƒ±sƒ± | `df.loc[1, "Num_Reviews"]` | `df.iloc[1, 3]` |

‚û°Ô∏è **Aynƒ± sonu√ß**, farklƒ± eri≈üim yolu.

---

#### Satƒ±r / S√ºtun Aralƒ±klarƒ±

```python
df.loc[0:5, ["Movie_Title", "Rating"]]
```

‚úî 5 **dahil**

##### `.iloc`**(hari√ß)** ile aralƒ±k

```python
df.iloc[0:5, 0:2]
```

‚ùå 5 **dahil deƒüil**

#### Filtreleme + .loc (EN √ñNEMLƒ∞ KISIM)

```python
df.loc[df["Rating"] >= 8, ["Movie_Title", "Rating"]]
```

üìå Bu kullanƒ±m:
- Profesyonel

#### Ne Zaman Hangisi?
‚úÖ `.loc` kullan:

- Filtreleme yapƒ±yorsan

- Notebook yazƒ±yorsan
