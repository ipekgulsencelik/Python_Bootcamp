# Pandas Merge (Join)

Bu notebook pandas `merge` konusunu örneklerle açıklar.

````python
import pandas as pd
````

## Örnek Veri

````python
customers = {
    'customer_id': [1, 2, 3],
    'full_name': ['Burak Yılmaz', 'Hakan Yılmaz', 'İpek Yılmaz'],
    'address': ['Ataköy', 'Sarıyer', 'Sarıyer']
}

orders = {
    'order_id': [1001, 1002, 1003, 1004, 1005],
    'customer_id': [1, 2, 3, 4, 5],
    'order_revenue': [2456, 1234, 4567, 2345, 9876]
}

df_customers = pd.DataFrame(customers)
df_orders = pd.DataFrame(orders)
````

````python
df_customers
````

````python
df_orders
````

## Merge

`merge`, iki veya daha fazla **DataFrame’i ortak bir anahtar (key)** üzerinden birleştirmeye yarar.
> SQL’deki `JOIN` işlemlerinin pandas karşılığıdır.

```python
pd.merge(left_df, right_df, on='key')
```

---

### 🔑 Temel Parametreler

| Parametre  | Açıklama                     |
| ---------- | ---------------------------- |
| `left`     | Sol DataFrame                |
| `right`    | Sağ DataFrame                |
| `on`       | Ortak kolon adı              |
| `how`      | Birleştirme türü             |
| `left_on`  | Sol DF kolon adı             |
| `right_on` | Sağ DF kolon adı             |
| `suffixes` | Aynı isimli kolonlar için ek |

---

### 🔗 Merge Türleri (how)

#### 1️⃣ Inner Merge (varsayılan)

**Inner merge**, iki DataFrame’i **ortak bir anahtar kolona göre** birleştirir ve
**SADECE her iki tabloda da bulunan ortak kayıtları** döndürür.

> Ortak olmayan kayıtlar **tamamen elenir.**

📌 **Kural**

> **Sol tabloda VAR + Sağ tabloda VAR → GELİR**
>
> **Biri yoksa → GELMEZ**

````python
# inner merge
inner_merge = pd.merge(
    left=df_customers,
    right=df_orders,
    on='customer_id',
    how='inner'
)

inner_merge
````

📌 Not:
- `how` parametresi yazılmadığında **varsayılan** = `inner`’dır.
- Yani bu kod zaten **inner join** yapar.

#### Left Merge

- **Ana tablo** (sol tablo)
- Çıktıdaki **satır sayısını belirleyen** tablo

````python
left_merge = pd.merge(
    left=df_orders,
    right=df_customers,
    on='customer_id',
    how='left'
)

left_merge
````

LEFT JOIN mantığı:

- Sol tablodaki (df_orders) **tüm satırlar korunur**
- Sağ tabloda eşleşme yoksa → `NaN`

#### Right Merge


````python
right_merge = pd.merge(
    left=df_customers,
    right=df_orders,
    on='customer_id',
    how='right'
)

right_merge
````
